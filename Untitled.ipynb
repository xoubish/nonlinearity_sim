```python
import astropy.io.fits as pyfits
from astropy.stats import sigma_clip

import matplotlib.pyplot as plt
import numpy as np
```

```python
dark = pyfits.getdata('sc456_crRejectSF.fits')
```

```python
np.shape(dark)[2]
```

```python
plt.imshow(dark[:,:,0])
```

```python
i,j = np.random.randint(2040),np.random.randint(2040)
b = sigma_clip(dark[i,j,:], sigma = 3)

plt.plot(np.arange(np.shape(dark)[2]),b)
```

```python
plt.plot(np.arange(np.shape(dark)[2]),np.cumsum(b))
plt.plot(np.arange(np.shape(dark)[2]),np.arange(np.shape(dark)[2])*1.48)
```
