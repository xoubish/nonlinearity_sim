```python
from matplotlib.colors import LogNorm

plt.title('Frame 165 (vmax = 40)')
plt.imshow(dark[:,:,165],vmax=40, origin = 'lower')
plt.colorbar()
```

```python
i,j = np.random.randint(2040),np.random.randint(2040)
b = sigma_clip(dark[i,j,:], sigma = 3)

plt.title('pixel:'+str(i)+','+str(j))
plt.plot(np.arange(np.shape(dark)[2]),b)
plt.xlabel('frames')
plt.ylabel('flux')
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

plt.title('pixel:'+str(i)+','+str(j))
plt.plot(np.arange(np.shape(dark)[2]),np.cumsum(b),label='cumulative flux')
plt.plot(np.arange(np.shape(dark)[2]),np.arange(np.shape(dark)[2])*1.48, label='1.48*frame')
#plt.plot(np.arange(np.shape(dark)[2]),np.cumsum(smooth(b,30)),label='smoothing')
plt.legend()
plt.xlabel('frames')
plt.ylabel('cumulative flux')
```
