```python
def sumRange(L,a,b,beta,printslopes):                                                                                                                                                                                                
    s = 0                                                                                                                                                                                                         
    for i in range(a,b):  
        if printslopes:
            if i == a:
                print('Group slopes:')
            print(i, ' ', L[i] - L[i-1])
        s += pow((L[i] - L[i-1]) + beta,2)  
    return s 

def make_random_trace(nelem=2e4, k=1., plots=0):
    ''' make a binary random trace in (-1,1) with exponentially distributed dwell 
    times. nelem = number of jumps. k = rate.'''
    k = float(k)
    dt = 1./k
    rg = np.random.rand( np.int(nelem) )
    ra = np.abs(rg)
    ra = ra/np.max(ra)
    # exponentially distributed jumps:
    t = np.log(1/ra)/k
    tr = np.floor(t/dt)
    # print('max trace ',np.max(tr))
    x = []
    s = 3.0e-3
    for j in np.arange(len(tr)):
        for ti in np.arange(np.int(tr[j])):
            x = np.append(x, s*np.ones(ti))
            s = -s
    time = np.arange(len(x))*dt
    spectrum_x = np.abs(np.fft.fft(x))**2
    freq_x = np.fft.fftfreq(len(x), d=dt)
    freq_x = freq_x[:int(len(spectrum_x)/2.)]
    spectrum_x = spectrum_x[:len(freq_x)]
    spectrum_x = spectrum_x/float(len(x)**2 *freq_x[1])
    spectrum_x[1:] = 2*spectrum_x[1:]

    if plots:
        plt.subplot(211)
        plt.plot(time[:], x[:], '-o')
        plt.subplot(212)
        plt.loglog(freq_x, spectrum_x, '.')
        plt.show()

    return x, time

def addRTN(RTN_num, fdarkN, nelements=2000,rate=0.2):
    ''' Gets the darkcurrent with no RTN and returns it with RTN if RTN_num is not 0.'''
    
    ru = np.random.uniform
    tf = 1.4548 # tf = 1.42 in DPU test Sec. 8.2.
    tread = np.arange(len(fdarkN))*tf
        
    if RTN_num==1: 
        sRTN, tRTN = make_random_trace(nelements,rate,0)  
        sRTNterp = ru(200)*np.interp(tread,tRTN,sRTN)
        fon = ru(0.0)
        if (fon < 0.5):
            fon = 0.5    #  At least 50%
        for RTN in range(0,nreads):
            on = ru(0.0)
            if (on <= fon):
                fdarkN[RTN] += sRTNterp[RTN]*np.mean(fdarkN)

    elif RTN_num==2:
        fon = ru(0.0) # frequency of readouts that should have RTN
        if (fon < 0.2):  
            fon = 0.2    #  At least 20%.
        level = 0.2*np.mean(fdarkN) # RTN maximum offset from nominal
        level = level*ru(0.0)
        for RTN in range(0,nreads):
            # Generate a random number to decide if this readout is telegraphing 
            # or not.
            on = ru(0.0) 
            if (on <= fon):
                fdarkN[RTN] += level
                
    elif RTN_num==3:
        sRTN, tRTN = make_random_trace(nelements,rate,0)
        sRTNterp = np.interp(tread,tRTN,sRTN)
        onoff = np.round(ru(np.float(len(fdarkN))))
        fdarkN = fdarkN + sRTNterp*0.15*np.mean(fdarkN)*onoff
    
    else:
        fdarkN = fdarkN

    return fdarkN

def nonlinear(fdark, nonlin,start=0.5):
    ''' Gets array of dark current adds non_linearity to it'''  
    
    u = np.int(start*len(fdark)) # 50% index
    for i in range(u,len(fdark)):
        fdark[i] = fdark[i]-(i-u)/len(fdark)*nonlin*(fdark[i]**2)
    return fdark
```

```python
def MACCsimulator(darkcurrent=0.04, nonlin=0.00005, readnoise =9.0, RTN_num=1, rtn_nelements=2000, rtn_rate=0.2 , plot='yes'):
    '''MACC simulator with inputs dark current, non-linearity, 
    readnoise and RTN returning the up the ramp DPU signal, 
    slope, quality flag'''
    
    lightcurrent, detGain, offsetADU = 10.0, 1.3, 0. #detGain = 1/1.4 DPU test; typos 8.4?
    ng,nf,nd=15, 16, 11
    tf = 1.4548 # tf = 1.42 in DPU test Sec. 8.2.
    
    sigmaRead = readnoise/detGain # ADU
    fluence = ( darkcurrent + lightcurrent )/ detGain  # ADU/sec
    tgrp = (nf + nd)*tf
    tint = (ng-1.)*(nf+nd)*tf
    nreads = int((ng*nf + (ng-1.)*nd))
    texp =  nreads*tf
    read = np.arange(nreads)
    tread = read*tf
    fdark = fluence*tread 
    
    #adding non_linearity
    fdarkN = nonlinear(fdark,nonlin,0.5) ## start nonlinearity halfway  
    if plot=='yes':
        plt.plot(tread,fdarkN,label='dark')
    
    #adding readnoise
    rg = np.random.rand(nreads)            
    fdarkN = fdarkN + rg * sigmaRead  / np.max(rg)  
    if plot=='yes':
        plt.plot(tread,fdarkN,label='+random noise')

    ## ADD RTN    
    fdarkN = addRTN(RTN_num, fdarkN ,nelements=rtn_nelements, rate=rtn_rate)
    
    
    ## MAAC group and measurements 
    alpha = (1 - (nf**2))/(3.*nf*(nf+nd))  
    beta = 2.*(sigmaRead**2)/(nf*(alpha+1))   
    s2,t2 = np.float(), np.float()
    for i in range(nreads):
        if i % int(nf+nd) in (range(nf)):
            s = fdarkN[i]
            t = tread[i]
            s2 = np.append(s2, s)
            t2 = np.append(t2, t) 
    
    if plot=='yes':
        plt.plot(t2,s2,'r.',label='groups')

    fgrp = np.reshape(s2[1:],[ng,nf])
    tgrp = np.reshape(t2[1:],[ng,nf])
    fgrpavg,tgrpavg =[],[]
    for i in range(ng):
        fga = np.mean(fgrp[i,:])
        tga = np.mean(tgrp[i,:])
        fgrpavg = np.append(fgrpavg, fga)
        tgrpavg = np.append(tgrpavg, tga)

    if plot=='yes':
        plt.plot(tgrpavg,fgrpavg,'g.',label='avg')
        plt.legend(loc=4)


    L = fgrpavg   
    deltaT = tgrpavg[1]-tgrpavg[0] 
    a = 1
    b = ng

    slopeDPU_1 = np.sqrt( 1 + 4 * ( sumRange(L,a,b,beta,0)/((ng - 1)*((1+alpha)**2)) ) )  # DPU Manual
    slopeDPU_0 = (1 + alpha) / 2.  # DPU manual
    slopeDPU = slopeDPU_0 * (slopeDPU_1 - 1) - beta
    signalDPU = slopeDPU * (ng - 1) + offsetADU # Downlinked signal in ADU
    pseudoFlux = np.sqrt ( sumRange(L,a,b,beta,0) / (ng - 1) ) - beta
    QF = ( 2*detGain/(1+alpha) ) * ( (ng - 1)*pseudoFlux - (fgrpavg[ng-1] - fgrpavg[0]) )

    # Simple least squares fit to the full ramp for comparison
    coeff = poly.polyfit(tread[1:], fdarkN[1:], 1)
    ffit = poly.polyval(tread[1:], coeff)
    chisqr = np.sum((poly.polyval(tread, coeff) - fdarkN) ** 2) / np.std(fdarkN)
    if plot =='yes':
        plt.plot(tread,poly.polyval(tread, coeff),'y--')

    return QF,signalDPU,slopeDPU,coeff[1]*(tgrpavg[-1]-tgrpavg[-2])
```

### MCMC over non-linearity and different types of RTN

```python
def mcmc_nonlinearity(steps=1500):
    '''MCMC over non_linearity and RTNs to see effect on QF'''
    
    noise,dark = 9.0, 0.04
    nl0 = 0.00002
    rtn0 = 0
    rtn_nelements0 = 2000
    rtn_rate0 = 0.2
    
    p0,s0,m0,ml0 = MACCsimulator(darkcurrent=dark, nonlin=nl0, readnoise =noise, RTN_num=rtn0, rtn_nelements=rtn_nelements0, rtn_rate=rtn_rate0 , plot='no')

    chain=[[nl0,rtn0,rtn_nelements0,rtn_rate0,p0,s0,m0,ml0]]
    x=[0]
    for i in range(steps):
    
        index=1#np.random.randint(1,2)
        nl1, rtn1, rtn_nelements1, rtn_rate1 = nl0, rtn0, rtn_nelements0, rtn_rate0
        nl1 = 0.00015*np.random.uniform(0,1)
        #if index==1:
        #    nl1 = nl0 + 0.00001*(np.random.uniform(-1,1))
        #    if (nl1 >= 0.0):
        #        nl1 = nl0 + 0.00001*(np.random.uniform(-1,0))
               
        rtn1 = np.random.randint(1,5)-1
        if rtn1 ==0:
            rtn_nelements1,rtn_rate1 = rtn_nelements0,rtn_rate0
        else:
            rtn_nelements1 = np.random.uniform(900,3000)
            rtn_rate1 = np.random.uniform(0.05,0.5)
  
        p0,s0,m0,ml0 = MACCsimulator(darkcurrent=dark, nonlin=nl0, readnoise =noise, RTN_num=rtn0, rtn_nelements=rtn_nelements0, rtn_rate=rtn_rate0 , plot='no')
        p1,s1,m1,ml1 = MACCsimulator(darkcurrent=dark, nonlin=nl1, readnoise =noise, RTN_num=rtn1, rtn_nelements=rtn_nelements1, rtn_rate=rtn_rate1 , plot='no')
        
        if i%1000==0:
            print(i)    
    
        if (np.abs(ml0-ml1) < 200):
            x.append(i)
            chain.append([nl1,rtn1,rtn_nelements1,rtn_rate1,p1,s1,m1,ml1])
            nl0, rtn0, rtn_nelements0, rtn_rate0  = nl1, rtn1, rtn_nelements1, rtn_rate1

    nevis = open('MAAC_params'+str(steps)+'.cat', 'w')
    for i in range(len(chain)):
        nevis.write(str(chain[i][0])+'\t'+str(chain[i][1])+'\t'+str(chain[i][2])+'\t'\
                    +str(chain[i][3])+'\t'+str(chain[i][4])+'\t'+str(chain[i][5])+'\t'+str(chain[i][6])+'\t'+str(chain[i][7])+'\n')
    nevis.close()   
    
    return chain
```

## Adding noise from the nghxrg 

```python
def white_noise(nstep=None):
    """
    Generate white noise for an HxRG including all time steps
    (actual pixels and overheads).
    Parameters:
    nstep - Length of vector returned
    """
    return(np.random.standard_normal(nstep))    

def pink_noise(mode,nstep):
    """
    Generate a vector of non-periodic pink noise.
    Parameters:
    mode - Selected from {'pink', 'acn'} """
        
    nstep2 = int(2**np.ceil(np.log2(nstep)))

    f1 = np.fft.rfftfreq(nstep2) # Frequencies for nstep elements
    f2 = np.fft.rfftfreq(2*nstep2) # ... for 2*nstep elements
    alpha = -1 # Hard code for 1/f noise until proven otherwise
    p_filter1 = np.sqrt(f1**alpha)
    p_filter2 = np.sqrt(f2**alpha)
    p_filter1[0] = 0.
    p_filter2[0] = 0.
   
    # Configure depending on mode setting
    if mode is 'pink':
        nstep  = 2*nstep
        nstep2 = 2*nstep2 # JML
        f = f2
        p_filter = p_filter2
    else:
        nstep  = nstep
        nstep2 = nstep2 # JML
        f = f1
        p_filter = p_filter1
    # Generate seed noise
    mynoise = white_noise(nstep2)
  
    # Save the mean and standard deviation of the first
    # half. These are restored later. We do not subtract the mean
    # here. This happens when we multiply the FFT by the pinkening
	# filter which has no power at f=0.
    the_mean = np.mean(mynoise[:nstep2//2])
    the_std = np.std(mynoise[:nstep2//2])
  
    # Apply the pinkening filter.
    thefft = np.fft.rfft(mynoise)
    thefft = np.multiply(thefft, p_filter)
    result = np.fft.irfft(thefft)
    result = result[:nstep//2] # Keep 1st half of nstep
    
    # Restore the mean and standard deviation
    result *= the_std / np.std(result)
    result = result - np.mean(result) + the_mean
 
    return(result)
```

### How is pink noise or ACN different from white noise?

```python
plt.figure(figsize=(6,4))
nsteps=10000
p = white_noise(nsteps)
a = plt.hist(p,histtype='step',color='white',linewidth=3,label='white',bins=20)

p = pink_noise('pink',nsteps)
a = plt.hist(p,histtype='step',color='pink',linewidth=3,bins=20,label='pink')
plt.legend(loc=1,fontsize=12)


p = pink_noise('acn',2*nsteps)
a = plt.hist(p,histtype='step',color='b',linewidth=3,bins=20,label='ACN')
plt.legend(loc=1,fontsize=12)
plt.ylabel('#',size=15)
plt.xlabel('noise',size=15)
```

Not different! Maybe the difference is only in the way its applied to the image? lets see

```python
w = 4 # Easier to work with

x1 = 0; y1 = 0
x2 = 2048; y2 = 2048
xsize = 2048

lower = w-y1; upper = w-(2048-y2)
left = w-x1; right = w-(2048-x2)
ref_all = np.array([lower,upper,left,right])
ref_all[ref_all<0] = 0

reference_pixel_noise_ratio = 0.8

naxis1,naxis2,naxis3 = 2048,2048,1  
```

```python
#read noise
w = ref_all
r = reference_pixel_noise_ratio  # Easier to work with

result = np.zeros((naxis3, naxis2, naxis1), dtype=np.float32)
for z in np.arange(1):
    here = np.zeros((naxis2, naxis1))
    # Noisy reference pixels for each side of detector
    if w[0] > 0: # lower
        here[:w[0],:] = r * rd_noise * np.random.standard_normal((w[0],naxis1))
    if w[1] > 0: # upper
        here[-w[1]:,:] = r * rd_noise * np.random.standard_normal((w[1],naxis1))
    if w[2] > 0: # left
        here[:,:w[2]] = r * rd_noise * np.random.standard_normal((naxis2,w[2]))
    if w[3] > 0: # right
        here[:,-w[3]:] = r * rd_noise * np.random.standard_normal((naxis2,w[3]))
    # Noisy regular pixels
    if np.sum(w) > 0: # Ref. pixels exist in frame
        here[w[0]:naxis2-w[1],w[2]:naxis1-w[3]] = rd_noise * np.random.standard_normal((naxis2-w[0]-w[1],naxis1-w[2]-w[3]))
    else: # No Ref. pixels, so add only regular pixels
        here = rd_noise * np.random.standard_normal((naxis2,naxis1))
    # Add the noise in to the result
    result[z,:,:] += here
```
